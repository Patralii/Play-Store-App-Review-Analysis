# **Project Summary -**

The Play Store App Review Analysis project aims to explore the Google Play Store applications dataset and extract meaningful business insights using Exploratory Data Analysis (EDA). The primary objective of this project is to understand the characteristics of apps available on the Google Play Store, identify patterns in user ratings and reviews, examine installation trends, and analyze factors that influence an application's popularity and user satisfaction.

The project began with extensive data preprocessing to ensure data quality. Missing values, duplicate records, and inconsistent data formats were identified and handled appropriately. Several columns such as Installs, Reviews, Price, Size, and Last Updated were cleaned and converted into suitable numerical or datetime formats to enable accurate analysis. This preprocessing stage ensured that the dataset was reliable for further exploration and visualization.

After cleaning the data, multiple visualizations were created to answer important business questions. Histograms were used to understand the distribution of app ratings, showing that most applications receive ratings between 4.0 and 4.5, indicating generally positive user feedback. Bar charts revealed that categories such as Family, Game, and Tools contain the highest number of applications, demonstrating where developers are most active.

Further analysis examined the relationship between numerical variables using a correlation heatmap. The results showed a moderate positive relationship between Reviews and Installs, suggesting that applications with higher user engagement generally attract more downloads. User review sentiment analysis indicated that the majority of reviews are positive, while negative reviews represent a smaller portion, highlighting an overall satisfactory user experience across the platform.

The analysis also compared free and paid applications, explored installation patterns across categories, and investigated relationships between app size, ratings, reviews, and installs. These findings provide valuable insights into developer strategies, user preferences, and market competition within the Google Play ecosystem.

From a business perspective, the project demonstrates that maintaining high app quality, encouraging positive user reviews, and focusing on high-demand categories can significantly improve application visibility and downloads. At the same time, categories with fewer applications may provide opportunities for developers to enter less competitive markets. Monitoring negative reviews can also help developers identify issues early and improve customer satisfaction through timely updates and bug fixes.

Overall, this project highlights how exploratory data analysis can transform raw app store data into actionable business insights. The findings can assist developers, product managers, and business stakeholders in making informed decisions regarding application development, pricing strategies, category selection, and user experience improvements. By leveraging data-driven insights, organizations can better understand customer behavior, improve product quality, and increase their chances of success in the competitive mobile application market.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
ps_df = ps_df.drop_duplicates()
ur_df = ur_df.drop_duplicates()
````

````python
ps_df.duplicated().sum()
ur_df.duplicated().sum()
````

````python
ps_df['Reviews'] = pd.to_numeric(ps_df['Reviews'], errors='coerce')
````

````python
ps_df['Reviews'].dtype
````

````python
# Remove ',' and '+' from Installs column
ps_df['Installs'] = ps_df['Installs'].astype(str).str.replace(',', '', regex=False)
ps_df['Installs'] = ps_df['Installs'].str.replace('+', '', regex=False)

# Convert to numeric
ps_df['Installs'] = pd.to_numeric(ps_df['Installs'], errors='coerce')

# Check datatype
ps_df['Installs'].dtype
````

````python
ps_df['Installs'].dtype
````

````python
# Convert to string first, then remove '$', then convert to numeric
ps_df['Price'] = ps_df['Price'].astype(str).str.replace('$', '', regex=False)

# Now convert to numeric
ps_df['Price'] = pd.to_numeric(ps_df['Price'], errors='coerce')

# Check datatype
print(ps_df['Price'].dtype)
````

````python
ps_df['Price'].head()
````

````python
ps_df['Size'] = ps_df['Size'].replace('Varies with device', np.nan)
````

````python
ps_df['Size'] = ps_df['Size'].astype(str).str.replace('k', 'e-3', regex=False)
ps_df['Size'] = ps_df['Size'].str.replace('M', '', regex=False)
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
````

````python
plt.figure(figsize=(8,5))

sns.histplot(ps_df['Rating'], bins=20, kde=True)

plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Apps')

plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram is suitable for understanding the distribution of a continuous numerical variable. It helps identify where most app ratings are concentrated and whether the ratings are normally distributed, skewed, or contain outliers.

##### 2. What is/are the insight(s) found from the chart?

Most applications have ratings between 4.0 and 4.5, indicating that a majority of apps on the Google Play Store receive favorable user feedback. Very few apps have ratings below 3, suggesting that poorly rated apps are relatively uncommon.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Business Impact:

Yes. Most apps have ratings between 4.0 and 4.5, indicating that users are generally satisfied with app quality. High-rated apps are more likely to attract new users, improve trust, increase downloads, and encourage positive word-of-mouth. Developers should continue improving user experience and fixing bugs to maintain high ratings.

Negative Business Impact:

The chart also shows a small number of apps with ratings below 3.0. These low-rated apps may experience fewer downloads, poor user retention, and negative reviews, which can reduce revenue and damage the app's reputation. Developers should identify the reasons behind low ratings and address user complaints through regular updates and feature improvements.

````python
# Count apps in each category
top_categories = ps_df['Category'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_categories.values,
    y=top_categories.index
)

plt.title('Top 10 Categories by Number of Apps')
plt.xlabel('Number of Apps')
plt.ylabel('Category')

plt.show()
````

The Family, Game, and Tools categories contain the highest number of applications on the Google Play Store, indicating that developers heavily focus on these categories due to high user demand. On the other hand, categories with fewer apps may represent niche markets with lower competition.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. The analysis helps businesses understand which app categories are highly competitive and which categories have fewer applications. Companies entering saturated categories like Family, Games, and Tools should differentiate their products through better features, quality, or pricing to stand out.

Negative Growth: Categories with a very high number of competing apps may experience slower growth for new entrants because users already have many alternatives. This intense competition can increase customer acquisition costs, reduce app visibility, and make it harder for new apps to gain downloads. Therefore, businesses should conduct additional market research before launching apps in overcrowded categories.

````python
# Chart - 5 visualization code
````

````python
plt.figure(figsize=(10,6))

avg_rating = (
    ps_df.groupby('Content Rating')['Rating']
    .mean()
    .sort_values(ascending=False)
)

sns.barplot(
    x=avg_rating.values,
    y=avg_rating.index
)

plt.title('Average Rating by Content Rating')
plt.xlabel('Average Rating')
plt.ylabel('Content Rating')

plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal bar chart is suitable for comparing the average ratings across different content rating categories. It clearly highlights which audience groups tend to receive higher user satisfaction.

##### 2. What is/are the insight(s) found from the chart?

Apps targeted toward specific content rating groups receive different average ratings. Categories such as Everyone and Everyone 10+ generally maintain consistently high ratings because they appeal to a broader audience. More restricted categories, such as Mature 17+, have fewer applications and may show greater variation in user ratings.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Business Impact:
Yes. Understanding how ratings differ across content categories helps developers identify target audiences with higher engagement and satisfaction. Businesses can optimize app content and marketing strategies to improve ratings and user retention.

Negative Growth:
Apps targeting niche or mature audiences may experience lower download potential and more varied ratings due to a smaller user base. Developers should focus on improving content quality and user experience to maintain strong ratings within these specialized categories.

#### Chart - 7

````python
# Chart - 7 visualization code
````

````python
plt.figure(figsize=(8,6))

sns.heatmap(
    ps_df[['Rating','Reviews','Size','Installs','Price']].corr(),
    annot=True,
    cmap='Greens',
    fmt='.2f'
)

plt.title('Correlation Heatmap of Numerical Features')
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is useful for understanding the strength and direction of relationships between multiple numerical variables at the same time. It quickly identifies which features are strongly correlated and which have little or no relationship.

##### 2. What is/are the insight(s) found from the chart?

The heatmap shows that Reviews and Installs have the strongest positive correlation (0.63), indicating that apps with more installs generally receive more user reviews. Other variables such as Rating, Size, and Price have very weak correlations with each other, suggesting they have limited direct influence on one another.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Business Impact:
Yes. The strong positive correlation between Installs and Reviews suggests that apps with more downloads receive more user feedback. Businesses can increase app visibility through marketing, ASO (App Store Optimization), and feature improvements to drive installs, collect more reviews, build credibility, and attract even more users.

Negative Business Impact:
The heatmap also shows that Rating has only a very weak correlation with Installs and Reviews, meaning that high ratings alone do not guarantee app growth. Similarly, Price has almost no relationship with other variables, indicating that pricing alone is not a major driver of downloads or user engagement. Businesses that focus only on ratings or pricing while ignoring marketing, user acquisition, and product quality may experience slower growth and reduced competitiveness.
